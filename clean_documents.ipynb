```python
import pandas as pd
import nltk
import jieba
import jieba.posseg as pseg

anshan_answers = pd.read_excel("知乎鞍山回答内容.xlsx")
anshan_questions = pd.read_excel("知乎鞍山话题.xlsx")
dalian_answers = pd.read_excel("知乎大连回答内容.xlsx")
dalian_questions = pd.read_excel("知乎大连话题.xlsx")
dongguan_answers = pd.read_excel("知乎东莞回答内容.xlsx")
dongguan_questions = pd.read_excel("知乎东莞话题.xlsx")
eerduosi_answers = pd.read_excel("知乎鄂尔多斯话题.xlsx")
eerduosi_questions = pd.read_excel("知乎鄂尔多斯回答内容.xlsx")
guangzhou_answers = pd.read_excel("知乎广州回答内容.xlsx")
guangzhou_questions = pd.read_excel("知乎广州话题.xlsx")
huhehaote_answers = pd.read_excel("知乎呼和浩特话题.xlsx")
huhehaote_questions = pd.read_excel("知乎呼和浩特回答内容.xlsx")
huizhou_answers = pd.read_excel("知乎惠州话题.xlsx")
huizhou_questions = pd.read_excel("知乎惠州回答内容.xlsx")
hulunbeier_answers = pd.read_excel("知乎呼伦贝尔话题.xlsx")
hulunbeier_questions = pd.read_excel("知乎呼伦贝尔回答内容.xlsx")
jinhua_answers = pd.read_excel("知乎金华回答内容.xlsx")
jinhua_questions = pd.read_excel("知乎金华话题.xlsx")
shenyang_answers = pd.read_excel("知乎沈阳回答内容.xlsx")
shenyang_questions = pd.read_excel("知乎沈阳话题.xlsx")
shenzhen_answers = pd.read_excel("知乎深圳回答内容.xlsx")
shenzhen_questions = pd.read_excel("知乎深圳话题.xlsx")
suzhou_answers = pd.read_excel("知乎苏州回答内容.xlsx")
suzhou_questions = pd.read_excel("知乎苏州话题.xlsx")
taiyuan_answers = pd.read_excel("知乎太原回答内容.xlsx")
taiyuan_questions = pd.read_excel("知乎太原话题.xlsx")

documents = [anshan_answers,
             anshan_questions,
             dalian_answers,
             dalian_questions,
             dongguan_answers,
             dongguan_questions,
             eerduosi_answers,
             eerduosi_questions,
             guangzhou_answers,
             guangzhou_questions,
             huhehaote_answers,
             huhehaote_questions,
             huizhou_answers,
             huizhou_questions,
             hulunbeier_answers,
             hulunbeier_questions,
             jinhua_answers,
             jinhua_questions,
             shenyang_answers,
             shenyang_questions,
             shenzhen_answers,
             shenzhen_questions,
             suzhou_answers,
             suzhou_questions,
             taiyuan_answers,
             taiyuan_questions]

for i in range(len(documents)):
    documents[i] = documents[i].iloc[:,0]
    documents[i] = documents[i].dropna()
```

```python
wordlist = pd.read_excel("CorpusWordlist.xls")

wordlist = wordlist.drop([0,1,2,3,4])
wordlist.columns = wordlist.iloc[0]
wordlist = wordlist.drop([5])
wordlist.reset_index(level=0, inplace=True)
wordlist = wordlist.drop(columns=["index"])

print(wordlist)

character_list = wordlist["词语"].tolist()
```

```python
full_text = []
full_text_no_tokenized = []
full_text_counts = []
full_text_ranked_diff = []

not_allowed = ["x", "eng"]

for doc in documents:
    text = doc.tolist()
    full_text_no_tokenized = full_text_no_tokenized + text
    tokenized_text = []
    for comment in text:
        tokenized = pseg.cut(comment)
        for word, flag in tokenized:
            if flag not in not_allowed:
                tokenized_text.append(word)
        
    full_text = full_text + tokenized_text
```

```python
with open("full_text.txt", "w") as f:
    for item in full_text_no_tokenized:
        f.write("%s\n" % item)
```

```python
anshan_questions_answers = anshan_questions.iloc[:,0].tolist() + anshan_answers.iloc[:,0].tolist()
anshan_fulltext = []
anshan_counts = []
anshan_ranked_diff = []

dalian_questions_answers = dalian_questions.iloc[:,0].tolist() + dalian_answers.iloc[:,0].tolist()
dalian_fulltext = []
dalian_counts = []
dalian_ranked_diff = []

dongguan_questions_answers = dongguan_questions.iloc[:,0].tolist() + dongguan_answers.iloc[:,0].tolist()
dongguan_fulltext = []
dongguan_counts = []
dongguan_ranked_diff = []

eerduosi_questions_answers = eerduosi_questions.iloc[:,0].tolist() + eerduosi_answers.iloc[:,0].tolist()
eerduosi_fulltext = []
eerduosi_counts = []
eerduosi_ranked_diff = []

guangzhou_questions_answers = guangzhou_questions.iloc[:,0].tolist() + guangzhou_answers.iloc[:,0].tolist()
guangzhou_fulltext = []
guangzhou_counts = []
guangzhou_ranked_diff = []

huhehaote_questions_answers =  huhehaote_questions.iloc[:,0].tolist() + huhehaote_answers.iloc[:,0].tolist()
huhehaote_fulltext = []
huhehaote_counts = []
huhehaote_ranked_diff = []

huizhou_questions_answers = huizhou_questions.iloc[:,0].tolist() + huizhou_answers.iloc[:,0].tolist()
huizhou_fulltext = []
huizhou_counts = []
huizhou_ranked_diff = []

hulunbeier_questions_answers = hulunbeier_questions.iloc[:,0].tolist() + hulunbeier_answers.iloc[:,0].tolist()
hulunbeier_fulltext = []
hulunbeier_counts = []
hulunbeier_ranked_diff = []

jinhua_questions_answers = jinhua_questions.iloc[:,0].tolist() + jinhua_answers.iloc[:,0].tolist()
jinhua_fulltext = []
jinhua_counts = []
jinhua_ranked_diff = []

shenyang_questions_answers = shenyang_questions.iloc[:,0].tolist() + shenyang_answers.iloc[:,0].tolist()
shenyang_fulltext = []
shenyang_counts = []
shenyang_ranked_diff = []

shenzhen_questions_answers = shenzhen_questions.iloc[:,0].tolist() + shenzhen_answers.iloc[:,0].tolist()
shenzhen_fulltext = []
shenzhen_counts = []
shenzhen_ranked_diff = []

suzhou_questions_answers = suzhou_questions.iloc[:,0].tolist() + suzhou_answers.iloc[:,0].tolist()
suzhou_fulltext = []
suzhou_counts = []
suzhou_ranked_diff = []

taiyuan_questions_answers = taiyuan_questions.iloc[:,0].tolist() + documents[24].tolist()
taiyuan_fulltext = []
taiyuan_counts = []
taiyuan_ranked_diff = []



questions_answers = [anshan_questions_answers, 
                     dalian_questions_answers, 
                     eerduosi_questions_answers, 
                     guangzhou_questions_answers,
                     huhehaote_questions_answers,
                     huizhou_questions_answers, 
                     hulunbeier_questions_answers, 
                     jinhua_questions_answers,
                     shenyang_questions_answers, 
                     shenzhen_questions_answers,
                     suzhou_questions_answers,
                     taiyuan_questions_answers]

fulltexts = [full_text,
             anshan_fulltext, 
             dalian_fulltext, 
             eerduosi_fulltext, 
             guangzhou_fulltext, 
             huhehaote_fulltext,
             huizhou_fulltext, 
             hulunbeier_fulltext, 
             jinhua_fulltext,
             shenyang_fulltext, 
             shenzhen_fulltext,
             suzhou_fulltext,
             taiyuan_fulltext]

counts = [full_text_counts,
          anshan_counts,
          dalian_counts,
          eerduosi_counts,
          guangzhou_counts,
          huhehaote_counts,
          huizhou_counts,
          hulunbeier_counts,
          jinhua_counts,
          shenyang_counts,
          shenzhen_counts,
          suzhou_counts,
          taiyuan_counts]

ranked_diff = [full_text_ranked_diff,
              anshan_ranked_diff,
              dalian_ranked_diff,
              eerduosi_ranked_diff,
              guangzhou_ranked_diff,
              huhehaote_ranked_diff,
              huizhou_ranked_diff,
              hulunbeier_ranked_diff,
              jinhua_ranked_diff,
              shenyang_ranked_diff,
              shenzhen_ranked_diff,
              suzhou_ranked_diff,
              taiyuan_ranked_diff]
          

not_allowed = ["x", "eng"]

for i in range(len(questions_answers)):
    for comment in questions_answers[i]:
        tokenized = pseg.cut(comment)
        for word, flag in tokenized:
            if flag not in not_allowed:
                fulltexts[i].append(word)
                
    print(fulltexts[i][:25])
```

```python
for i in range(len(counts)):
    counts[i] = pd.Series(fulltexts[i])
    counts[i] = counts[i].value_counts()
    counts[i] = counts[i].rename_axis('character').to_frame('counts')
    counts[i].reset_index(level=0, inplace=True)
    counts[i]["rank"] = counts[i]["counts"].rank(axis = 0, ascending=False)
    counts[i].reset_index(inplace=True)
    counts[i] = counts[i].drop(columns=["index"])

    for index, row in counts[i].iterrows():
        word = row["character"]
        if word not in character_list:
            counts[i] = counts[i].drop(index)
        
    counts[i].reset_index(inplace=True)
    counts[i] = counts[i].drop(columns=["index"])
    
    counts[i]["wordlist_rank"] = ""
    counts[i]["difference"] = ""


    for index, row in counts[i].iterrows():
        word = row["character"]
        word_rank = row["rank"]
        wordlist_rank = wordlist[wordlist['词语'] == word].index.item()
        wordlist_rank = wordlist_rank + 1
        difference = wordlist_rank - word_rank
        counts[i].iloc[index,3] = wordlist_rank
        counts[i].iloc[index,4] = difference
    
    print(counts[i][:10])
```

```python
for i in range(len(ranked_diff)):
    ranked_diff[i] = counts[i]
    ranked_diff[i].sort_values("difference", inplace=True, ascending=False)
    print(ranked_diff[i][:100])
```
